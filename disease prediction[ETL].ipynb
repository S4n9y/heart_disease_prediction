```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
print(df.head(5))
```

```python
text_cols = [
    'gender',
    'smoking',
    'alcohol_consumption',
    'physical_activity',
    'family_history',
    'disease'
]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["bmi_category"] = df["bmi"].apply(bmi_category)
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Middle"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(age_group)
```

```python
print(df.head(5))
```

```python
yes_no = [
    'smoking',
    'alcohol_consumption',
    'family_history',
    'disease'
]

for col in yes_no:
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
df.to_csv("cleaned_heart_disease_data.csv", index=False)

saved_df = pd.read_csv("cleaned_heart_disease_data.csv")
print(saved_df.head(5))
```
